```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 1
s = 0.3
f = 0.2
x0 = 2
t0 = 0

def dxdt(k,s,f,x):
    return k*x - s*x**2 - f

def euler(t0,x0,h,n):
    t = np.zeros(n+1)
    x = np.zeros(n+1)
    x[0] = x0
    t[0] = t0
    for i in range(n):
        x[i+1] = x[i] + h*dxdt(k,s,f,x[i])
        t[i+1] = t[i] + h
    return t,x
```

```python
t,x = euler(t0,x0,0.1,100)
plt.plot(t,x, label="Με ψάρεμα")
f = 0 
t,x = euler(t0,x0,0.1,100)
plt.plot(t,x, label="Χωρίς Ψάρεμα")

f = k**2/(4*s)
t,x = euler(t0,x0,0.1,100)
plt.plot(t,x, label="Κρίσιμος ρυθμός")
plt.grid()

plt.title("Πληθυσμός Ψαριών")
plt.legend()
```
